### 2.2. Convertimos tipos de datos 

```python
print(df.dtypes)
```

```python
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce').astype('Int64')

df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce')
```

Se realizaron las siguientes conversiones:

• year_of_release:
Convertida de tipo objeto a entero (Int64). Esto permite realizar análisis cronológicos y agrupar los juegos por año. Se usó errors=coerce para convertir valores no numéricos en NaN.

• user_score:
Convertida de objeto a flotante (float64). Esta columna contenía valores como TBD, que no representan puntuaciones reales. Se usó errors=coerce para convertir esos valores en NaN.

• critic_score:
Convertida de objeto a flotante (float64) para permitir cálculos estadísticos y comparaciones con otras variables numéricas.

### 2.3. Tratamos los valores ausentes

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['year_of_release'])
```

```python
df['rating'] = df['rating'].fillna('unknown')
```

Se identificaron valores nulos en varias columnas:

• year_of_release:
Se eliminaron las filas con valores ausentes, ya que el año de lanzamiento es esencial para el análisis temporal y no puede ser estimado con precisión.

• rating:
Se rellenaron los valores nulos con unknown, ya que algunos juegos no tienen clasificación ESRB (por ejemplo, títulos independientes o lanzamientos internacionales). Esta estrategia permite mantener la categoría sin distorsionar el análisis.

• user_score y critic_score:
Se dejaron los valores nulos como NaN. Rellenarlos con promedios podría introducir sesgos, especialmente si el juego no tiene reseñas por falta de popularidad o por ser reciente.

Podemos decir que las posibles causas de los valores ausentes serían :

• Falta de reseñas: Juegos nuevos o poco populares pueden no tener puntuaciones de usuarios o críticos.

• Errores de registro: Algunos datos pueden haberse perdido durante la recopilación.

• Ausencia de clasificación ESRB: No todos los juegos son evaluados por esta entidad, especialmente los que no se distribuyen en EE.UU.

# 3. Análisis de datos

### 3.1. Distribución de juegos por año

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_clean = df.copy()

df_clean['year_of_release'] = pd.to_numeric(df_clean['year_of_release'], errors='coerce')

df_clean = df_clean.dropna(subset=['year_of_release'])

df_clean['year_of_release'] = df_clean['year_of_release'].astype(int)

df_clean = df_clean.sort_values(by='year_of_release')

sns.set(style="whitegrid")

plt.figure(figsize=(14, 6))
sns.countplot(x='year_of_release', data=df_clean, palette='viridis')
plt.xticks(rotation=90)
plt.title('Cantidad de Juegos por Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Juegos')
plt.tight_layout()
plt.show()
```

Se realizó una limpieza de la columna year_of_release, eliminando las filas con valores nulos y convirtiendo los datos a tipo entero, esto nos permitió generar un gráfico claro y ordenado de la cantidad de juegos lanzados por año.El análisis muestra que la mayoría de los lanzamientos se concentran entre 2005 y 2015, con un pico notable en 2008. A partir de 2012 se observa una ligera disminución en la cantidad de títulos publicados. Esta tendencia puede estar relacionada con cambios en la industria, como el auge de plataformas digitales o la consolidación de franquicias exitosas.

### 3.8. Comparación de ventas del mismo juego en otras plataformas

Se seleccionaron los cinco juegos multiplataforma con mayores ventas globales y se compararon sus ventas en cada plataforma donde fueron lanzados.
Se observa que títulos como Call of Duty o FIFA tienden a vender más en consolas con mayor base de usuarios, como PS3 y X360, aunque algunos juegos también tienen buen rendimiento en PC o Wii.

# 5. Pruebas de hipótesis

### Hipótesis 1: Xbox One vs PC

• Hipótesis nula (H₀): Las calificaciones promedio de los usuarios para Xbox One y PC son iguales.
• Hipótesis alternativa (H₁): Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.

```python
from scipy import stats

xbox = df[(df['platform'] == 'XOne') & (df['user_score'].notna())]['user_score']
pc = df[(df['platform'] == 'PC') & (df['user_score'].notna())]['user_score']

t_stat1, p_val1 = stats.ttest_ind(xbox, pc, equal_var=False)

print(f"T-statistic: {t_stat1:.2f}")
print(f"P-value: {p_val1:.4f}")
```

### Hipótesis 2: Acción vs Deportes

• Hipótesis nula (H₀): Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son iguales.
• Hipótesis alternativa (H₁): Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes.

```python
action = df[(df['genre'] == 'Action') & (df['user_score'].notna())]['user_score']
sports = df[(df['genre'] == 'Sports') & (df['user_score'].notna())]['user_score']

t_stat2, p_val2 = stats.ttest_ind(action, sports, equal_var=False)

print(f"T-statistic: {t_stat2:.2f}")
print(f"P-value: {p_val2:.4f}")
```

Para ambas pruebas se utilizó la prueba t de Student para muestras independientes, con un umbral de significancia de α = 0.05.• Si el p-valor < 0.05, se rechaza la hipótesis nula y se concluye que hay una diferencia significativa.
• Si el p-valor ≥ 0.05, no se rechaza la hipótesis nula, lo que indica que no hay evidencia suficiente para afirmar que las medias son diferentes.

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es *estadísticamente diferente*. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).
